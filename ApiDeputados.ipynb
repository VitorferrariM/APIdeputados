```python
import requests
import pandas as pd
from time import sleep

# IDs e nomes dos 10 deputados
deputados = {
    220714: "Adail Filho",
    204560: "Adolfo Viana",
    220538: "Albuquerque",
    204501: "Alencar Santana",
    221148: "Alfredinho",
    220576: "Alfredo Gaspar",
    226708: "Allan Garc√™s",
    220676: "Andreia Siqueira",
    123756: "Ant√¥nia L√∫cia",
    74090:  "√Åtila Lins"
}

# Lista para guardar todas as proposi√ß√µes
todas_proposicoes = []

url_props = "https://dadosabertos.camara.leg.br/api/v2/proposicoes"

for id_dep, nome in deputados.items():
    print(f"üîé Buscando proposi√ß√µes de {nome}...")
    pagina = 1
    while True:
        params = {
            "idDeputadoAutor": id_dep,
            "itens": 100,
            "pagina": pagina,
            "ordem": "DESC",
            "ordenarPor": "ano"
        }
        r = requests.get(url_props, params=params)
        if r.status_code != 200:
            print(f"‚ö†Ô∏è Erro ao buscar proposi√ß√µes de {nome}")
            break

        dados = r.json().get("dados", [])
        if not dados:
            break

        for prop in dados:
            todas_proposicoes.append({
                "Deputado": nome,
                "Tipo": prop.get("siglaTipo"),
                "N√∫mero": prop.get("numero"),
                "Ano": prop.get("ano"),
                "Ementa": prop.get("ementa", "")
            })

        pagina += 1
        sleep(0.2)

# Converter para DataFrame
df_todas = pd.DataFrame(todas_proposicoes)

# üîé Filtrar apenas proposi√ß√µes que N√ÉO s√£o PL
df_outros = df_todas[df_todas["Tipo"] != "PL"]

# Mostrar por deputado
# Mostrar proposi√ß√µes (sem PL) separadas por deputado
for dep in deputados.values():
    df_dep = df_outros[df_outros["Deputado"] == dep]
    if not df_dep.empty:
        print(f"\nüìú Proposi√ß√µes de {dep} (exceto PL):")
        for _, row in df_dep.iterrows():
            print(f" - {row['Tipo']} {row['N√∫mero']}/{row['Ano']}: {row['Ementa']}")
    else:
        print(f"\n‚ÑπÔ∏è {dep} n√£o possui proposi√ß√µes diferentes de PL registradas.")
```

No c√≥digo desenvolvido, foi feita uma integra√ß√£o com a API de Dados Abertos da C√¢mara dos Deputados para coletar informa√ß√µes sobre proposi√ß√µes apresentadas por um grupo espec√≠fico de dez parlamentares. A l√≥gica percorre cada deputado, acessa as p√°ginas da API e armazena em uma tabela (DataFrame do Pandas) os dados principais de cada proposi√ß√£o: tipo (REQ, INC, PEC, PL etc.), n√∫mero, ano e ementa, que √© o resumo ou ‚Äúnome‚Äù da proposta.

Ap√≥s a coleta, o c√≥digo permite filtrar os resultados de acordo com o interesse da an√°lise. Por exemplo, √© poss√≠vel visualizar apenas os Projetos de Lei (PL) ou, ao contr√°rio, excluir os PL e observar apenas outros tipos de proposi√ß√µes, como Requerimentos (REQ), Indica√ß√µes (INC) ou Propostas de Emenda √† Constitui√ß√£o (PEC). Al√©m disso, os resultados podem ser organizados por deputado, facilitando a compara√ß√£o entre eles.

Essas informa√ß√µes s√£o √∫teis porque permitem acompanhar a atua√ß√£o parlamentar de forma transparente e estruturada. Ao listar as proposi√ß√µes, conseguimos identificar:

```python
import requests
import pandas as pd
def get_deputado_eventos(deputado_id):
    url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{deputado_id}/eventos"
    response = requests.get(url, params={"itens": 100})  # Pega at√© 100 eventos por p√°gina
    if response.status_code == 200:
        data = response.json()
        return data['dados']
    else:
        print("Erro ao acessar API:", response.status_code)
        return []
```

```python
def calcular_frequencia(deputado_id):
    eventos = get_deputado_eventos(deputado_id)
    
    if not eventos:
        return 0
    
    # Considerando que cada evento √© uma sess√£o, ent√£o frequ√™ncia √© n√∫mero de eventos
    total_eventos = len(eventos)
    
    # Aqui voc√™ poderia pegar total de sess√µes realizadas, se quiser propor√ß√£o
    # Para simplifica√ß√£o, apenas retorna total de presen√ßas
    return total_eventos
```

```python
def get_deputados():
    url = "https://dadosabertos.camara.leg.br/api/v2/deputados"
    response = requests.get(url, params={"itens": 10})
    if response.status_code == 200:
        data = response.json()
        deputados = data['dados']
        return deputados
    else:
        print("Erro ao acessar API:", response.status_code)
        return []

deputados = get_deputados()
print(f"Total de deputados encontrados: {len(deputados)}")
```

```python
def get_deputado_eventos(deputado_id):
    url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{deputado_id}/eventos"
    response = requests.get(url, params={"itens": 100})  # Pega at√© 100 eventos por p√°gina
    if response.status_code == 200:
        data = response.json()
        return data['dados']
    else:
        return []
```

```python
frequencias = []

for deputado in deputados:
    deputado_id = deputado['id']
    nome = deputado['nome']
    
    eventos = get_deputado_eventos(deputado_id)
    frequencias.append({
        "id": deputado_id,
        "nome": nome,
        "presencas": len(eventos)
    })
    
    

df_frequencias = pd.DataFrame(frequencias)
df_frequencias = df_frequencias.sort_values(by="presencas", ascending=False)
df_frequencias.head(10)
```

```python
import matplotlib.pyplot as plt

# Top 10 deputados
top10 = df_frequencias.head(10)

plt.figure(figsize=(12,6))
plt.bar(top10['nome'], top10['presencas'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Deputados por N√∫mero de Presen√ßas')
plt.xlabel('Deputados')
plt.ylabel('N√∫mero de Presen√ßas')
plt.show()
```
