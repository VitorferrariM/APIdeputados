```python
import requests
import pandas as pd
from time import sleep

# IDs e nomes dos 10 deputados
deputados = {
    220714: "Adail Filho",
    204560: "Adolfo Viana",
    220538: "Albuquerque",
    204501: "Alencar Santana",
    221148: "Alfredinho",
    220576: "Alfredo Gaspar",
    226708: "Allan Garcês",
    220676: "Andreia Siqueira",
    123756: "Antônia Lúcia",
    74090:  "Átila Lins"
}

# Lista para guardar todas as proposições
todas_proposicoes = []

url_props = "https://dadosabertos.camara.leg.br/api/v2/proposicoes"

for id_dep, nome in deputados.items():
    print(f"🔎 Buscando proposições de {nome}...")
    pagina = 1
    while True:
        params = {
            "idDeputadoAutor": id_dep,
            "itens": 100,
            "pagina": pagina,
            "ordem": "DESC",
            "ordenarPor": "ano"
        }
        r = requests.get(url_props, params=params)
        if r.status_code != 200:
            print(f"⚠️ Erro ao buscar proposições de {nome}")
            break

        dados = r.json().get("dados", [])
        if not dados:
            break

        for prop in dados:
            todas_proposicoes.append({
                "Deputado": nome,
                "Tipo": prop.get("siglaTipo"),
                "Número": prop.get("numero"),
                "Ano": prop.get("ano"),
                "Ementa": prop.get("ementa", "")
            })

        pagina += 1
        sleep(0.2)

# Converter para DataFrame
df_todas = pd.DataFrame(todas_proposicoes)

# 🔎 Filtrar apenas proposições que NÃO são PL
df_outros = df_todas[df_todas["Tipo"] != "PL"]

# Mostrar por deputado
# Mostrar proposições (sem PL) separadas por deputado
for dep in deputados.values():
    df_dep = df_outros[df_outros["Deputado"] == dep]
    if not df_dep.empty:
        print(f"\n📜 Proposições de {dep} (exceto PL):")
        for _, row in df_dep.iterrows():
            print(f" - {row['Tipo']} {row['Número']}/{row['Ano']}: {row['Ementa']}")
    else:
        print(f"\nℹ️ {dep} não possui proposições diferentes de PL registradas.")
```

No código desenvolvido, foi feita uma integração com a API de Dados Abertos da Câmara dos Deputados para coletar informações sobre proposições apresentadas por um grupo específico de dez parlamentares. A lógica percorre cada deputado, acessa as páginas da API e armazena em uma tabela (DataFrame do Pandas) os dados principais de cada proposição: tipo (REQ, INC, PEC, PL etc.), número, ano e ementa, que é o resumo ou “nome” da proposta.

Após a coleta, o código permite filtrar os resultados de acordo com o interesse da análise. Por exemplo, é possível visualizar apenas os Projetos de Lei (PL) ou, ao contrário, excluir os PL e observar apenas outros tipos de proposições, como Requerimentos (REQ), Indicações (INC) ou Propostas de Emenda à Constituição (PEC). Além disso, os resultados podem ser organizados por deputado, facilitando a comparação entre eles.

Essas informações são úteis porque permitem acompanhar a atuação parlamentar de forma transparente e estruturada. Ao listar as proposições, conseguimos identificar:

```python
import requests
import pandas as pd
def get_deputado_eventos(deputado_id):
    url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{deputado_id}/eventos"
    response = requests.get(url, params={"itens": 100})  # Pega até 100 eventos por página
    if response.status_code == 200:
        data = response.json()
        return data['dados']
    else:
        print("Erro ao acessar API:", response.status_code)
        return []
```

```python
def calcular_frequencia(deputado_id):
    eventos = get_deputado_eventos(deputado_id)
    
    if not eventos:
        return 0
    
    # Considerando que cada evento é uma sessão, então frequência é número de eventos
    total_eventos = len(eventos)
    
    # Aqui você poderia pegar total de sessões realizadas, se quiser proporção
    # Para simplificação, apenas retorna total de presenças
    return total_eventos
```

```python
def get_deputados():
    url = "https://dadosabertos.camara.leg.br/api/v2/deputados"
    response = requests.get(url, params={"itens": 10})
    if response.status_code == 200:
        data = response.json()
        deputados = data['dados']
        return deputados
    else:
        print("Erro ao acessar API:", response.status_code)
        return []

deputados = get_deputados()
print(f"Total de deputados encontrados: {len(deputados)}")
```

```python
def get_deputado_eventos(deputado_id):
    url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{deputado_id}/eventos"
    response = requests.get(url, params={"itens": 100})  # Pega até 100 eventos por página
    if response.status_code == 200:
        data = response.json()
        return data['dados']
    else:
        return []
```

```python
frequencias = []

for deputado in deputados:
    deputado_id = deputado['id']
    nome = deputado['nome']
    
    eventos = get_deputado_eventos(deputado_id)
    frequencias.append({
        "id": deputado_id,
        "nome": nome,
        "presencas": len(eventos)
    })
    
    

df_frequencias = pd.DataFrame(frequencias)
df_frequencias = df_frequencias.sort_values(by="presencas", ascending=False)
df_frequencias.head(10)
```

```python
import matplotlib.pyplot as plt

# Top 10 deputados
top10 = df_frequencias.head(10)

plt.figure(figsize=(12,6))
plt.bar(top10['nome'], top10['presencas'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Deputados por Número de Presenças')
plt.xlabel('Deputados')
plt.ylabel('Número de Presenças')
plt.show()
```
